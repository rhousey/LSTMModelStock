# LSTM Model Stock Price Prediction AMZN

## Imports 

```python
from datetime import date, timedelta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import yfinance as yf
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.models import Sequential
```

```python
data = yf.download('AMZN', start=(date.today() - timedelta(days=5000)),
                   end=date.today() - timedelta(days=30))
data["Date"] = data.index
```

## Filter by Close Price

```python
data
```

```python
close_data = data["Close"].values.reshape(-1,1)
train_data, test_data = train_test_split(close_data, test_size = 0.15 , shuffle=False)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_train_data = scaler.fit_transform(train_data)
scaled_test_data = scaler.fit_transform(test_data)
```

```python
x_train = []
y_train = []
for i in range(100, train_data.shape[0]):
    x_train.append(scaled_train_data[i - 100:i])
    y_train.append(scaled_train_data[i,0])
```

```python
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1],1))
x_train.shape
```

## Building LSTM Model 


```python
model = Sequential()
# increasing layers, units tuning  
model.add(LSTM(units = 256, return_sequences = True, input_shape= (x_train.shape[1], 1)))
model.add(Dropout(0.2))

model.add(LSTM(units = 128, return_sequences = True))
model.add(Dropout(0.2))

model.add(LSTM(units = 64))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.summary()
```

```python
import tensorflow as tf 

model.compile(optimizer = 'adam' , loss = 'mean_squared_error', metrics = [tf.keras.metrics.MeanAbsoluteError()])
# changed from epochs = 20, batch_size = 128 
model.fit(x_train, y_train, epochs = 50, batch_size = 64)
```

```python
X_test = []
y_test = []
for i in range(100,len(scaled_test_data)):
    X_test.append(scaled_test_data[i - 100:i, 0])
    y_test.append(scaled_test_data[i, 0])

X_test, y_test = np.array(X_test), np.array(y_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Predict using the trained model
y_pred = model.predict(X_test)

scale = 1 / scaler.scale_[0]
y_pred = y_pred * scale
y_test = y_test * scale
```

```python
test_dates = data.index[-len(scaled_test_data)+100:]
test_dates
```

```python
plt.figure(figsize=(28,12))
plt.plot(test_dates, y_test, color = 'blue', label = 'Price')
plt.plot(test_dates, y_pred, color = 'red', label = 'Predicted Price')
plt.xlabel('Time')
plt.title('Stock Price and Predicted Price')
plt.legend()
plt.show()
```

## Future Price Prediction of next day

```python
newData = yf.download('AMZN', start=(date.today() - timedelta(days=45)),
                   end=date.today())
```

```python
new_df = newData.filter(['Close'])
last30Days = new_df[-30:].values
last30Days_scaled = scaler.transform(last30Days)
X_test_data = [] 
X_test_data.append(last30Days_scaled)
X_test_data = np.array(X_test_data)
X_test_data = np.reshape(X_test_data, (X_test_data.shape[0], X_test_data.shape[1], 1))
```

```python
predicted_price = model.predict(X_test_data)
```

```python
predicted_price = scaler.inverse_transform(predicted_price)
```
